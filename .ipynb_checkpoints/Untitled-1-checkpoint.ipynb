```python
import pandas as pd
import numpy as np



caminho = 'C:/Users/maria.nommad/Desktop/conda/dados_financeiros.xlsx'
dados = pd.read_excel(caminho)
dados['Data_ref'] = pd.to_datetime(dados['Data_ref'])
dados

```

```python
# Calcular o Patrimônio total por mês

patrimonio_por_mes = dados.groupby(dados['Data_ref'].dt.to_period('M'))['Patrimonio'].sum()
patrimonio_por_mes
```

```python
top3_maiores_patrimonios = patrimonio_por_mes.nlargest(3)
top3_maiores_patrimonios
```

```python
top3_menores_patrimonios = patrimonio_por_mes.nsmallest(3)
top3_menores_patrimonios
```

```python
# Supondo que 'dados' seja o DataFrame com as colunas 'Prospeccao', 'Atendimento' e 'Taxa_adm'

# Função para calcular o repasse
def calcular_repasse(row):
    if row['Prospeccao'] == row['Atendimento']:
        return row['Taxa_adm'] * 0.8
    else:
        if row['Prospeccao'] != row['Atendimento']:
            if row['Prospeccao'] == 2 or row['Atendimento'] == 2:
                return row['Taxa_adm'] * 0.6
            else:
                return row['Taxa_adm'] * 0.2
        else:
            return 0

# Aplicando a função para calcular o repasse e criando a coluna 'Repasse'
dados['Repasse'] = dados.apply(calcular_repasse, axis=1)
dados
```

```python
def calcular_repasse(row):
    if row['Produto'] == 1:
        if row['Prospeccao'] == row['Atendimento']:
            return row['Taxa_adm'] * 0.6  # 60% para Prospeccao e Atendimento iguais para Produto 1
        else:
            return row['Taxa_adm'] * 0.45 + row['Taxa_adm'] * 0.15  # 45% para Prospeccao e 15% para Atendimento diferentes para Produto 1
    else:
        if row['Prospeccao'] == row['Atendimento']:
            return row['Taxa_adm'] * 0.8  # 80% para Prospeccao e Atendimento iguais para outros Produtos
        else:
            return row['Taxa_adm'] * 0.6  # 60% para Prospeccao e 20% para Atendimento diferentes para outros Produtos
```

```python
# Aplicando a função para calcular o repasse e criando a coluna 'Repasse'
dados['Repasse'] = dados.apply(calcular_repasse, axis=1)

# Exibindo o DataFrame com a nova coluna 'Repasse'
dados
```

```python
# Supondo que 'dados' seja o DataFrame que contém as informações
# Use o groupby para calcular o total do patrimônio por segmento
patrimonio_por_segmento = dados.groupby('Segmento')['Patrimonio'].sum()

# Encontre o segmento com o maior patrimônio
segmento_maior_patrimonio = patrimonio_por_segmento.idxmax()
maior_patrimonio = patrimonio_por_segmento.max()

print(f"Segmento com o maior patrimônio: {segmento_maior_patrimonio}")
print(f"Maior patrimônio: {maior_patrimonio}")
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Supondo que 'dados' seja o DataFrame com a coluna 'Data_ref' representando as datas
# Converta a coluna 'Data_ref' para o tipo datetime, se ainda não estiver no formato correto
dados['Data_ref'] = pd.to_datetime(dados['Data_ref'])

# Crie uma coluna com o mês extraído da coluna 'Data_ref'
dados['Mes'] = dados['Data_ref'].dt.to_period('M')

# Calcule a quantidade de clientes únicos por mês
clientes_por_mes = dados.groupby('Mes')['Cliente'].nunique()

# Crie um gráfico de linha para representar a evolução da quantidade de clientes por mês
plt.figure(figsize=(10, 6))
clientes_por_mes.plot(kind='line', marker='o', color='blue')
plt.title('Evolução da Quantidade de Clientes Mensais')
plt.xlabel('Mês')
plt.ylabel('Quantidade de Clientes')
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
